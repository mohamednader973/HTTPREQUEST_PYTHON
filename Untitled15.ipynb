HTTP REQUESTS AND REST APIs

```python
import requests 
url='https://www.ibm.com/'
r=requests.get(url)
r.status_code
```

```python
header=r.headers
```

```python
r.request.body
```

```python
header['date']
```

```python
r.encoding
```

```python
r.text
```

GET & POST requests

```python
url_get='http://httpbin.org/get'
payload={"name":"joseph","ID":"123"}
rg=requests.get(url_get,params=payload)
```

```python
rg.url
```

```python
rg.status_code
```

```python
rg.text
```

```python
rg.json()
```

```python
rg.json()['headers']
```

```python
url_post='http://httpbin.org/post'

rp=requests.post(url_post,data=payload)
```

```python
rp.url
```
